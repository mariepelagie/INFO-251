### 1.2 Training and testing

Chances are, for the above problem you used all of your data to fit the regression line. In some circumstances this is a reasonable thing to do, but often this will result in overfitting. Let's redo the above results the ML way, using careful cross-validation.  Since you are now experts in cross-validation, and have written your own cross-validation algorithm from scratch, you can now take a shortcut and use the libraries that others have built for you.

Using the [cross-validation functions](http://scikit-learn.org/stable/modules/cross_validation.html) from scikit-learn, use 25-fold cross-validation to fit regression (a) above, i.e. the linear fit of housing price on number of rooms per house. Each fold of cross-validation will give you one slope coefficient and one intercept coefficient.  Plot the distribution of the 25 slope coefficients using a histogram, then draw a vertical line at the value of the slope coefficient that you estimated in 1.1 using the full dataset.  What do you notice?

```python
from sklearn.model_selection import KFold
# Data for Multivariate Model 
X = bdata["RM"].values
y = bdata["MEDV"].values

# number of folds in CV , K-folds
k = 25
# Initialize iteration counter and results list
iter_num= 0
Results = [0]*k
# creating k-folds=25
kf = KFold(n_splits = k,shuffle = True)

for train, test in kf.split(X):
    X_train, X_test, y_train, y_test = X[train], X[test], y[train], y[test]
    # Iteration number
    print("------------------------- Iteration", iter_num + 1, "-------------------------")

    # Linear regression model
    lr_rm = LinearRegression()

    # Reshape X for sklearn format
    X_train = np.reshape(X_train, (-1, 1))
    X_test = np.reshape(X_test, (-1, 1))

    # Fit the model and save interceps and coefficient in result list 
    lr_rm.fit(X_train, y_train)
    Results[iter_num] = [lr_rm.intercept_, lr_rm.coef_[0]]

    
    # Print results
    print("Linear Regression Model: MEDV ~ RM")
    print("Number of coefficients:", len(lr_rm.coef_))
    print("---------------- Model Coefficients --------------------")
    print("Intercept value:", lr_rm.intercept_)
    print("RM:               ", lr_rm.coef_[0])
    print("------------------------------------------------------")
    print("RMSE:", np.sqrt(np.mean((y_test - lr_rm.predict(X_test)) ** 2)))
    print("R-Squared:" , lr_rm.score(X_train, y_train))
    print("----------------------------------------------------------------")

    # Next iteration
    iter_num += 1
```

```python
#  Plot the distribution of the 25 slope coefficients using a histogram, 
# then draw a vertical line at the value of the slope coefficient 
# that you estimated in 1.1 using the full dataset. 

# Slopes
#Slopes = [Results[i][1] for i in range(0, k)]

# Figure size
plt.figure(figsize = (9, 5)) 

# Plot distribution of cefficients
plt.hist([Results[i][1] for i in range(0, k)], color = 'g', alpha = 0.5)
plt.axvline(Slope, color = 'k', linestyle = 'dashed', 
            linewidth = 3, alpha = 0.5)

# Title and labels
plt.title('Distribution of the 25 slope coefficients')
plt.xlabel('Coefficient value [1000$/NÂ° of rooms]')
plt.ylabel('Frequency');

# Double check mean values 
print("Mean value slopes (25-fold CV):", np.mean([Results[i][1] for i in range(0, k)]))
print("Original slope:                ", Slope)
```

```python
## Further Analysis, using the quadratic model
# Data for Multivariate Model 
X = np.array(bdata[["RM", "RM2"]])
y = np.array(bdata["MEDV"])

# number of folds in CV , K-folds
k = 25
# Initialize iteration counter and results list
iter_num= 0
Results = [0]*k
# creating k-folds=25
kf = KFold(n_splits = k,shuffle = True)

for train, test in kf.split(X):
    X_train, X_test, y_train, y_test = X[train], X[test], y[train], y[test]
    # Iteration number
    print("------------------------- Iteration", iter_num + 1, "-------------------------")

    # Linear regression model
    lr_rm2 = LinearRegression()


    # Fit the model and save interceps and coefficient in result list 
    lr_rm2.fit(X_train, y_train)
    Results[iter_num] = [lr_rm2.intercept_, lr_rm2.coef_[0], lr_rm2.coef_[1]]

    
    # Print results
    print("Linear Regression Model: MEDV ~ RM")
    print("Number of coefficients:", len(lr_rm2.coef_))
    print("---------------- Model Coefficients --------------------")
    print("Intercept value:", lr_rm2.intercept_)
    print("RM:               ", lr_rm2.coef_[0])
    print("------------------------------------------------------")
    print("RMSE:", np.sqrt(np.mean((y_test - lr_rm2.predict(X_test)) ** 2)))
    print("R-Squared:" , lr_rm2.score(X_train, y_train))
    print("----------------------------------------------------------------")

    # Next iteration
    iter_num += 1
```

```python
import time

"""
Function
--------
bivariate_ols
    Gradient Decent to minimize OLS. Used to find co-efficients of bivariate OLS Linear regression

Parameters
----------
xvalues, yvalues : narray
    xvalues: independent variable
    yvalues: dependent variable
    
R: float
    Learning rate
    
MaxIterations: Int
    maximum number of iterations
    

Returns
-------
alpha: float
    intercept
    
beta: float
    co-efficient
"""
def bivariate_ols(xvalues, yvalues, R=0.01, MaxIterations=1000):
    # start timer
    start_time = time.time()

    ## initialize alpha and beta
    alpha = 0
    beta = 0
    ## initialize alphas_vec and betas_vec to track value changes over time
    alphas_vec = np.zeros(MaxIterations + 1)
    betas_vec = np.zeros(MaxIterations + 1)
    # initialise results vector to store loss 

    results = np.zeros(MaxIterations + 1)
    ii = 0
    while True :
        # loss function
        temp = alpha + (beta * xvalues) - yvalues
        loss = 0.5 * np.mean(np.power(temp, 2))
        results[ii] = loss
        
        # Calculate partial derivatives
        grad_alpha = np.mean(temp)
        grad_beta= np.mean(temp * xvalues)
        
        
        ## Update alpha and beta values and store them
        alpha -= R* grad_alpha
        beta -= R* grad_beta
        alphas_vec[ii] = alpha
        betas_vec[ii] = beta
        
        ## Stoping criteria
        # Convergence of loss function value
        if np.abs(results[ii] - results[ii-1]) < 1e-10 and ii > 1:
            print("Convergence of Loss function reached, gradient descent terminates")
            break
        
        # Check convergence of parameters (after first iteration)
        if np.abs(alphas_vec[ii]  - alphas_vec[ii-1] ) < 1e-10  \
           and np.abs(betas_vec[ii]  - betas_vec[ii-1] ) < 1e-10 \
           and iters > 1: 
            print("Convergence of parameteres, alpha and beta reached, gradient descent terminates")
            break
        
        # Exceeding maximum number of iterations
        if ii + 1 > MaxIterations:
            print("Max number of iterations reached, gradient descent terminates")
            break
        ## increment counter 
        ii += 1
        
        
    

    print("Time taken: {:.2f} seconds".format(time.time() - start_time))
    print("Number of iteration: ", ii)
    return alpha, beta

# example function call
# Model Data
X = bdata["RM"]
Y = bdata["MEDV"]
print(bivariate_ols(X, Y, 0.01, 100000))
```

```python
"""
Function
--------
multivariate_ols
    Gradient Decent to minimize OLS. Used to find co-efficients of bivariate OLS Linear regression

Parameters
----------
xvalue_matrix, yvalues : narray
    xvalue_matrix: independent variable
    yvalues: dependent variable
    
R: float
    Learning rate
    
MaxIterations: Int
    maximum number of iterations
    

Returns
-------
alpha: float
    intercept
    
beta_array: array[float]
    co-efficient
"""

def multivariate_ols(xvalue_matrix, yvalues, R=0.01, MaxIterations=100000,Intercept = True ,intial_guess =[]):
    # start timer
    start_time = time.time()
    
    # If intercept True, add extra column of ones to feature matris
    if Intercept:
        intercept = np.ones((xvalue_matrix.shape[0], 1))
        x = np.hstack((intercept, xvalue_matrix))
        
    # Initialize betas and cost function vectors    
    betas = np.zeros(x.shape[1])
    results = np.zeros(MaxIterations+1)
    #betas_vec = [[0 for i in xvalue_matrix.columns] for i in range(0, MaxIterations + 1)]
    
    # If initial guess for betas
    if len(intial_guess) > 0:
        betas = InitialGuess

    ii = 0   
    # Main loop
    while ii<= MaxIterations:
        # loss function
        y_hat = np.dot(x,betas)
        error =  y_hat - yvalues
        results[ii]= 0.5 * np.mean(np.power(error, 2))
        
        # Update betas with gradient
        gradient = np.dot(x.T, error) / x.shape[0]
        betas -= R * gradient
       
        ## Stoping criteria
        # Convergence of loss function value
        if np.abs(results[ii] - results[ii-1]) < 1e-10 and ii > 1:
            print("Convergence of Loss function reached, gradient descent terminates")
            break
        
        # Exceeding maximum number of iterations
        if ii + 1 > MaxIterations:
            print("Max number of iterations reached, gradient descent terminates")
            break
            
        # if no decrease in objective( no improvement)
        if (results[ii] - results[ii-1]) > 0 and ii > 0:
            print("No improvement, break")
            break
        ## increment counter 
        ii += 1
        
        
    

    print("Time taken: {:.2f} seconds".format(time.time() - start_time))
    print("Total number of iterations:", ii)
    alpha = betas[0]
    beta_array = betas[1:]
    return alpha, beta_array
```

### **2.5 Implement  Mini-Batch Gradient Descent (MB-GD)**
MB-GD is a Gradient Descent variant that in large data sets can converge faster and is computationally less intensive. Implement MB-GD for question 2.3. Tune the learning rate, number of iterations and "mini-batch" size
so that you compute the estimates within a 1e-2 tolerance. Do not use a batch-size greater than 32.
In summary you go over the entire data set for n epochs. At the beginning of each epoch you shuffle your data once and then you pick k batches (approximately k=#of data points/batch_size). For each batch you compute the gradient, update the parameters and move to the next batch.
MB-GD is similar to Stochastic Gradient Descent but instead of using one sample to compute the gradient we use a batch of samples at each iteration. You can find details about MB-GD here:
https://en.wikipedia.org/wiki/Stochastic_gradient_descent

```python
# Implement Mini-Batch Gradient Descent (MB-GD)
"""
Function
--------
mini_batch
    Mini-Batch Gradient Decent to minimize OLS. Used to find co-efficients of bivariate OLS Linear regression

Parameters
----------
xvalue_matrix, yvalues : narray
    xvalue_matrix: independent variable
    yvalues: dependent variable
    
R: float
    Learning rate
    
n_epochs: Int
    maximum number of iterations
    
batch_size: int
    size of the batch tandomly selected for gradient calculation
    
intial_guess: list
    list containing initial guess for betas

Returns
-------
alpha: float
    intercept
    
beta_array: array[float]
    co-efficient
"""
def mini_batch(xvalue_matrix, yvalues,batch_size,n_epochs,R=0.01, Intercept = True ,intial_guess =[], shuffle= False,verbose=False):
    # start timer
    start_time = time.time()
    
    # If intercept True, add extra column of ones to feature matris
    if Intercept:
        intercept = np.ones((xvalue_matrix.shape[0], 1))
        x = np.hstack((intercept, xvalue_matrix))
    ## cap batch size to 32
    if batch_size>=32:
        batch_size = 32
        
    # Initialize betas and cost function vectors    
    betas = np.zeros((x.shape[1],1))
    loss_list =[]
    #betas_vec = [[0 for i in xvalue_matrix.columns] for i in range(0, MaxIterations + 1)]
    
    # If initial guess for betas
    if len(intial_guess) > 0:
        betas = InitialGuess

      
    # Main loop
    ii = 0
    #for ii in range(n_epochs): 
    while ii <=n_epochs:
        # create mini-batches
        mini_batches = [] 
        data = np.hstack((x, yvalues)) 
        np.random.shuffle(data) 
        n_minibatches = data.shape[0] // batch_size 
        i = 0

        for i in range(n_minibatches + 1): 
            mini_batch = data[i * batch_size:(i + 1)*batch_size, :] 
            X_mini = mini_batch[:, :-1] 
            Y_mini = mini_batch[:, -1].reshape((-1, 1)) 
            mini_batches.append((X_mini, Y_mini)) 
        if data.shape[0] % batch_size != 0: 
            mini_batch = data[i * batch_size:data.shape[0]] 
            X_mini = mini_batch[:, :-1] 
            Y_mini = mini_batch[:, -1].reshape((-1, 1)) 
            mini_batches.append((X_mini, Y_mini)) 
        loss=[]    
        for j,mini_batch in enumerate(mini_batches): 
            X_mini, Y_mini = mini_batch 
            y_hat = np.dot(X_mini,betas)
            y_hat = y_hat.reshape((len(Y_mini), 1))
            if verbose == True:
                print('y_hat.shape',y_hat.shape)
                print('Y_mini.shape',Y_mini.shape)
                print('X_mini.T.shape',X_mini.T.shape)
            error =  y_hat - Y_mini
            if verbose == True:
                print('error.shape',error.shape)
            # Update betas with gradient
            gradient = np.dot(X_mini.T, error)
            if verbose == True:
                print('gradient.shape',gradient.shape)
                print('betas.shape',betas.shape)
            betas -= R * gradient
            # Update loss function
            loss.append( 0.5 * np.mean(np.power(error, 2)))
             # Convergence of loss function value
            if np.abs(loss[j] - loss[j-1]) < 1e-2 and ii > 1:
                print("Convergence of Loss function reached, gradient descent terminates")
                print("Time taken: {:.2f} seconds".format(time.time() - start_time))
                print("Total number of iterations:", ii)
                alpha = betas[0]
                beta_array = betas[1:]
                return alpha, beta_array
#                 break
        loss_list.append(loss)

        ## Stoping criteria
        # Convergence of loss function value
        if np.abs(loss_list[ii][j] - loss_list[ii-1][j]) < 1e-2 and ii > 1:
            print("Convergence of Loss function reached, gradient descent terminates")
            break
            
        ## increment counter 
        ii += 1
    print("Maximum iteration limit exceeded, gradient descent terminates")
 

    print("Time taken: {:.2f} seconds".format(time.time() - start_time))
    print("Total number of iterations:", ii)
    alpha = betas[0]
    beta_array = betas[1:]
    return alpha, beta_array
```

```python
## Implement logistic regression.
## Useful functions
# Sigmoid function
def sig_function(exponent):
    return 1 / (1 + np.exp(- exponent))

# Log likelihood function (Cost function)
def LL(x, y, betas):
    exponent = np.dot(x, betas)
    ll = np.sum( y * exponent - np.log(1 + np.exp(exponent)) )
    return ll
```

```python
# Logistic Regression using gradient descent 
def logistic_regression(x, y, max_iter, R,  Intercept = False):
    # Starting time
    start_time = time.time()
    
    # If intercept is true, add it to the matrix
    if Intercept:
        intercept = np.ones((x.shape[0], 1))
        x = np.hstack((intercept, x))
        
    # Initialize betas and cost function    
    betas = np.zeros(x.shape[1])
    loss_function = np.zeros(max_iter)
    
    # Iterations
    for ii in range(max_iter):
        temp = np.dot(x, betas)
        y_hat = sig_function(temp)

        # Update weights with gradient
        error = y - y_hat
        gradient = np.dot(x.T, error)
        betas += R * gradient
        
        # Compute ll
        loss_function[ii] = LL(x, y, betas)
        
    
        # Check convergence
        if np.abs(loss_function[ii] - loss_function[ii - 1]) < 1e-10 and ii > 0:
            print("Convergence reached (Cost function), break")
            break
            
        # Check if no improvement
        if -(loss_function[ii] - loss_function[ii - 1]) > 0 and ii > 0:
            print("No improvement, break")
            break
        
    # Finalization   
    print("Time taken: " + str(round(time.time() - start_time,2)) + " seconds")
    print("Total number of iterations:", ii)
    
    # Return betas array
    return betas
```

### 4.3 Ridge regularization
Incorporate L2 (Ridge) regularization into your multivariate_ols regression. Write a new version of your gradient descent algorithm that includes a regularization term "lambda" to penalize excessive complexity. 

Use your regularized regression to re-fit the model from 4.2 above on your training data. Try this for several different values of lambda, and report your RMSE for each lambda separately for your training and testing data.
```python
# Your code here
"""
Function
--------
Ridge_reg
    Gradient Decent to minimize regularized regressions. 
    Used to find co-efficients of  multivariate Linear regression with Ridge regularization

Parameters
----------
xvalue_matrix, yvalues : narray
    xvalue_matrix: independent variable
    yvalues: dependent variable
    
R: float
    Learning rate
    
MaxIterations: Int
    maximum number of iterations
    

Returns
-------
alpha: float
    intercept
    
beta_array: array[float]
    co-efficient
"""

def Ridge_reg(xvalue_matrix, yvalues,Lambda, R=0.001, MaxIterations=100000,Intercept = True ,intial_guess =[]):
    # start timer
    start_time = time.time()
    
    # If intercept True, add extra column of ones to feature matris
    if Intercept:
        intercept = np.ones((xvalue_matrix.shape[0], 1))
        x = np.hstack((intercept, xvalue_matrix))
        
    # Initialize betas and cost function vectors    
    betas = np.zeros(x.shape[1])
    Lambdas = np.repeat(Lambda, x.shape[0])
    Lambdas[0] = 0.0
    lam = np.repeat(Lambda, x.shape[1])
    lam[0]=0
    ## no penalty on intercept
    results = np.zeros(MaxIterations+1)
    
    # If initial guess for betas
    if len(intial_guess) > 0:
        betas = InitialGuess

    ii = 0   
    # Main loop
    while ii<= MaxIterations:
        # loss function
        y_hat = np.dot(x,betas)
        error =  y_hat - yvalues
        results[ii]= 0.5 * (np.mean(np.power(error, 2))) +  np.dot(lam, np.power(betas, 2))
        
        # Update betas with gradient
        gradient = np.dot(x.T, error)  + 2 * np.dot(x.T, Lambdas) / x.shape[0]
        betas -= R * gradient
       
        ## Stoping criteria
        # Convergence of loss function value
        if np.abs(results[ii] - results[ii-1]) < 1e-10 and ii > 1:
            print("Convergence of Loss function reached, gradient descent terminates")
            break
        
        # Exceeding maximum number of iterations
        if ii + 1 > MaxIterations:
            print("Max number of iterations reached, gradient descent terminates")
            break
            
        # if no decrease in objective( no improvement)
        if (results[ii] - results[ii-1]) > 0 and ii > 0:
            print("No improvement, break")
            break
        ## increment counter 
        ii += 1
        
        
    

    print("Time taken: {:.2f} seconds".format(time.time() - start_time))
    print("Total number of iterations:", ii)
    alpha = betas[0]
    beta_array = betas[1:]
    return alpha, beta_array
```

```python
# Adaptive gradient descent (AdaGrad)

        
# def adagrad(xvalue_matrix, yvalues,args,stepsize = 1e-2,max_iterations=100000, batch_size=None, batch_ratio = 0.1):
#     gti=np.zeros(x0.shape[0])
    
#     #ld=len(data)
#     if batch_size is None:
#         minibatchsize = int(math.ceil(len(xvalue_matrix)*batch_ratio))
#     w=x0
#     for t in xrange(max_iterations):
#         s=sample(xrange(ld),minibatchsize)
#         sd=[data[idx] for idx in s]
#         grad=f_grad(w,sd,*args)
#         gti+=grad**2
#         adjusted_grad = grad / (fudge_factor + np.sqrt(gti))
#         w = w - stepsize*adjusted_grad
#     return w

"""
Function
--------
adagrad
    Adaptive Gradient Decent to on single sample or batches where stepsizes are adpated for each feature   
    in proportion to its gradient

Parameters
----------
xvalue_matrix, yvalues : narray
    xvalue_matrix: independent variable
    yvalues: dependent variable
    
R: float
    Learning rate
    
MaxIterations: Int
    maximum number of iterations
    
fudge_factor: float
    small number to counter numerical instabiltiy
    
batch_size: int
    number of training samples used for  iteration
    
verbose: boolean
    Flag to print extra information  
    

Returns
-------
alpha: float
    intercept
    
beta_array: array[float]
    co-efficient
"""

def adagrad(xvalue_matrix, yvalues,x0 = [],R = 0.01,max_iterations=100000, batch_size=1, verbose = False,shuffle=True, fudge_factor=1e-6):
    # start timer
    start_time = time.time()
    ## to use later to compute adjusted gradient
    
    # If intercept True, add extra column of ones to feature matris
    if Intercept:
        intercept = np.ones((xvalue_matrix.shape[0], 1))
        x = np.hstack((intercept, xvalue_matrix))
    else:
        x = xvalue_matrix
    temp=np.zeros((x.shape[1],1))
    
#       # set batch size if none specified
#     if batch_size is None:
#         minibatchsize = int(np.ceil(len(xvalue_matrix)*batch_ratio))
         
    # Initialize betas and cost function vectors    
    betas = np.zeros((x.shape[1],1))
    loss_list =[]
    #betas_vec = [[0 for i in xvalue_matrix.columns] for i in range(0, MaxIterations + 1)]
    
    # If initial guess for betas
    if len(x0) > 0:
        betas = InitialGuess

      
    # Main loop
    ii = 0
    #for ii in range(n_epochs): 
    while ii <=max_iterations:
        # create mini-batches
        mini_batches = [] 
        data = np.hstack((x, yvalues)) 
        np.random.shuffle(data) 
        n_minibatches = data.shape[0] // batch_size 
        i = 0

        for i in range(n_minibatches + 1): 
            mini_batch = data[i * batch_size:(i + 1)*batch_size, :] 
            X_mini = mini_batch[:, :-1] 
            Y_mini = mini_batch[:, -1].reshape((-1, 1)) 
            mini_batches.append((X_mini, Y_mini)) 
        if data.shape[0] % batch_size != 0: 
            mini_batch = data[i * batch_size:data.shape[0]] 
            X_mini = mini_batch[:, :-1] 
            Y_mini = mini_batch[:, -1].reshape((-1, 1)) 
            mini_batches.append((X_mini, Y_mini)) 
            
        loss=[]    
        for j,mini_batch in enumerate(mini_batches): 
            X_mini, Y_mini = mini_batch 
            y_hat = np.dot(X_mini,betas)
            y_hat = y_hat.reshape((len(Y_mini), 1))
            if verbose == True:
                print('y_hat.shape',y_hat.shape)
                print('Y_mini.shape',Y_mini.shape)
                print('X_mini.T.shape',X_mini.T.shape)
            error =  y_hat - Y_mini
            if verbose == True:
                print('error.shape',error.shape)
            # compute gradient and adjusted 
            gradient = np.dot(X_mini.T, error)
            if verbose == True:
                print('gradient.shape',gradient.shape)
            temp+= gradient**2
            if verbose == True:
                print('temp.shape',temp.shape)
            adjusted_grad = gradient / (fudge_factor + np.sqrt(temp))
            if verbose == True:
                print('gradient.shape',gradient.shape)
                print('betas.shape',betas.shape)
            # Update betas with adjusted gradient
            betas -= R * adjusted_grad
            # Update loss function
            loss.append( 0.5 * np.mean(np.power(error, 2)))
             # Convergence of loss function value
            if np.abs(loss[j] - loss[j-1]) < 1e-2 and ii > 1:
                print("Convergence of Loss function reached, gradient descent terminates")
                print("Time taken: {:.2f} seconds".format(time.time() - start_time))
                print("Total number of iterations:", ii)
                alpha = betas[0]
                beta_array = betas[1:]
                return alpha, beta_array
#                 break
        loss_list.append(loss)

        ## Stoping criteria
        # Convergence of loss function value
        if np.abs(loss_list[ii][j] - loss_list[ii-1][j]) < 1e-2 and ii > 1:
            print("Convergence of Loss function reached, gradient descent terminates")
            break
            
        ## increment counter 
        ii += 1
    print("Maximum iteration limit exceeded, gradient descent terminates")
 

    print("Time taken: {:.2f} seconds".format(time.time() - start_time))
    print("Total number of iterations:", ii)
    alpha = betas[0]
    beta_array = betas[1:]
    return alpha, beta_array
    
```
